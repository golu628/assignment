Q1. What is MongoDB? Explain non-relational databases in short. When is MongoDB preferred over SQL?
MongoDB is a popular NoSQL (non-relational), open-source database that stores data in JSON-like documents (BSON).

Non-relational databases store data without fixed table schemas. Instead of rows and columns, they use flexible structures like:

Key-value pairs

Documents

Graphs

Wide-columns

When to prefer MongoDB over SQL:

When dealing with large-scale unstructured data.

If data schema changes frequently.

When you need faster reads/writes and horizontal scaling.

For real-time analytics, IoT apps, or content management systems.

Q2. Features of MongoDB
Document-Oriented: Stores data in BSON format (binary JSON).

Schema-less: Each document can have a different structure.

High Performance: Fast reads/writes.

Scalable: Supports horizontal scaling via sharding.

Rich Query Language: Supports filters, projections, aggregation.

Indexing: Supports indexing for performance.

Aggregation Framework: Used for data analysis.

Replication: Provides high availability with replica sets.

Copy

    {"name": "Bob", "age": 24},
    {"name": "Charlie", "age": 23}

record = collection.find_one({"name": "Alice"})
print("find_one():", record)

# Find all
print("find():")

Q5. Using find() method to query MongoDB
The find() method retrieves documents that match a given condition.

python
Copy
Edit
# Find students aged more than 22
query = {"age": {"$gt": 22}}

results = collection.find(query)

print("Students with age > 22:")
for doc in results:
    print(doc)
