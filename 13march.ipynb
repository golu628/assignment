Q1. ANOVA Assumptions and Violations

ANOVA relies on several assumptions for accurate results:

Normality of Residuals: The errors (residuals) should be normally distributed. Violations can occur due to skewed data or outliers.
Homogeneity of Variances: The variances of the groups being compared should be equal. Heteroscedasticity (unequal variances) can lead to inflated Type I error rates.
Independence of Errors: Observations within and between groups should be independent. Dependence (e.g., repeated measures on the same subjects) requires specific ANOVA variations.
Q2. Types of ANOVA

One-Way ANOVA: Compares means across one independent variable (factor) with two or more levels (groups).
Two-Way ANOVA: Analyzes the effects of two independent variables on a dependent variable.
Repeated Measures ANOVA: Assesses the impact of a single independent variable with multiple levels measured on the same subjects.
Q3. Partitioning of Variance

ANOVA partitions the total variance (SST) into two components:

Explained Sum of Squares (SSE): Represents the variance attributable to the independent variable(s).
Residual Sum of Squares (SSR): Reflects the unexplained variance due to error or random factors.
Understanding this division helps determine how much of the total variability can be explained by the independent variable(s).

Q4. Calculating Sums of Squares in One-Way ANOVA (Python)

Here's a Python code snippet for calculating sums of squares:

Python
import numpy as np

def calculate_anova_ss(data, groups):
  n_total = len(data)
  group_means = np.mean(data, axis=0)
  overall_mean = np.mean(data)

  sst = np.sum((data - overall_mean) ** 2)
  sse = np.sum([(data[i] - group_means[groups[i]]) ** 2 for i in range(n_total)])
  ssr = sst - sse

  return sst, sse, ssr

Q7. Missing Data in Repeated Measures ANOVA

Missing data handling methods in repeated measures ANOVA include:

Listwise Deletion: Removes cases with missing values, potentially reducing power.
Mean/Median Imputation: Replaces missing values with group means/medians, potentially introducing bias.
Model-Based Methods (EM algorithm): More sophisticated approaches, but require specific software and expertise.
The choice depends on the amount and pattern of missing data, and each method has its trade-offs.

Q8. Post-Hoc Tests after ANOVA

Common post-hoc tests:

Tukey's HSD (Honestly Significant Difference): Compares all possible pairs of means, adjusting for multiple comparisons.
Scheffe's Test: More conservative than Tukey's HSD, suitable for unbalanced designs.
Bonferroni Correction: A simple method, but can be too conservative, reducing power.
Use a post-hoc test when ANOVA indicates significant overall effects to identify specific group differences. Example: If weight loss results in ANOVA are significant, a post-hoc test helps determine which diet(s) differ from others.
