Q1. Confidence Interval for Mean

Python
from scipy import stats

# Sample mean and standard deviation
mean = 50
std_dev = 5

# Confidence level (95%)
confidence_level = 0.95

# Calculate critical value (z-score)
z_crit = stats.norm.ppf(1 + (confidence_level / 2))

# Calculate confidence interval bounds
lower_bound = mean - (z_crit * std_dev)
upper_bound = mean + (z_crit * std_dev)

# Print results
print("Confidence Interval (95%):", lower_bound, "-", upper_bound)

# Interpretation
print("We can be 95% confident that the true population mean falls between", lower_bound, "and", upper_bound)

Q4. Confidence Interval for Smoking Proportion

from scipy import stats

# Sample size and number of smokers
n = 500
successes = 60

# Confidence level (95%)
confidence_level = 0.95

# Calculate margin of error
margin_of_error = stats.norm.ppf(1 + (confidence_level / 2)) * np.sqrt((successes / n) * (1 - (successes / n)))

# Confidence interval bounds
lower_bound = (successes / n) - margin_of_error
upper_bound = (successes / n) + margin_of_error

# Print results
print("Confidence Interval (95%):", lower_bound, "-", upper_bound)

# Interpretation
print("We can be 95% confident that the true proportion of smokers in the population falls between", lower_bound, "and", upper_bound)

Q6. Plotting Chi-Square Distribution

Python
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Degrees of freedom
df = 10

# Chi-square values for the x-axis
chi_sq_values = np.linspace(0, 30, 100)

# Probability density function (PDF) of chi-square distribution
pdf = chi2.pdf(chi_sq_values, df)

# Plot the PDF
plt.plot(chi_sq_values, pdf, label="Chi-Square Distribution (df=10)")

# Shade the area above chi-square statistic of 15
plt.fill_between(chi_sq_values, 0, pdf, where=chi_sq_values >= 15, alpha=0.2, color='red', label="Area > 15")

# Labels and title
plt.xlabel("Chi-Square Statistic")
plt.ylabel("Probability Density")
plt.title("Chi-Square Distribution with Shaded Area (chi^2 > 15)")
plt.legend()

# Display the plot
plt.show()
