# Q1. Class and Object

# A class is a blueprint or template for creating objects. It defines the structure and behavior that the objects will have.  Think of it like a cookie cutter.

# An object is an instance of a class. It's a concrete realization of the blueprint.  Think of the cookies you make using the cookie cutter.  Each cookie is a separate object.

class Dog:  # Class (Blueprint)
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        return "Woof!"

my_dog = Dog("Buddy", "Golden Retriever")  # Object (Instance)
your_dog = Dog("Max", "German Shepherd")  # Another Object (Instance)

print(my_dog.name)  # Accessing object's attribute
print(your_dog.bark())  # Calling object's method


 Q2. Four Pillars of OOPs

 1. Abstraction: Hiding complex implementation details and showing only essential information to the user.
 2. Encapsulation: Bundling data (attributes) and methods (behavior) that operate on that data within a class, and controlling access to them.
 3. Inheritance: Creating new classes (derived classes) from existing classes (base classes), inheriting their attributes and methods.
  4. Polymorphism: The ability of an object to take on many forms, allowing methods to be used with objects of different classes.

  Q3. __init__() function

 The __init__() function (also called the constructor) is a special method in Python classes that is automatically called when an object of the class is created. It's used to initialize the object's attributes (data).

class Person:
    def __init__(self, name, age):
        self.name = name  # Initializing the name attribute
        self.age = age    # Initializing the age attribute

person1 = Person("Alice", 30)  # __init__() is called here
print(person1.name)
print(person1.age)

 Q4. Why self is used

ans- `self` is a convention (though strongly recommended) in Python used to refer to the instance of the class within the class methods. When you call a method on an object, Python implicitly passes the object itself as the first argument to the method.  `self` is used to access and modify the object's attributes.

class Car:
  def __init__(self, model):
    self.model = model  # self refers to the current Car object

  def display_model(self):
    print("The model of this car is", self.model) # self is used to access the object's attribute

my_car = Car("Toyota")
my_car.display_model() # Python implicitly passes my_car to the display_model method


 Q5. Inheritance and its types

 Inheritance allows you to create new classes (derived classes or subclasses) that inherit attributes and methods from existing classes (base classes or superclasses). This promotes code reuse and creates a hierarchy of classes.

 Types of Inheritance:

1. Single Inheritance: A derived class inherits from only one base class.

class Animal:
    def speak(self):
        print("Generic animal sound")

class Dog(Animal): # Dog inherits from Animal
    def bark(self):
        print("Woof!")

my_dog = Dog()
my_dog.speak()  # Inherited from Animal
my_dog.bark()

 2. Multiple Inheritance: A derived class inherits from multiple base classes.

class Flyable:
    def fly(self):
        print("Can fly")

class Swimmable:
    def swim(self):
        print("Can swim")

class FlyingFish(Flyable, Swimmable): # FlyingFish inherits from both Flyable and Swimmable
    pass

fish = FlyingFish()
fish.fly()
fish.swim()

 3. Multilevel Inheritance: A derived class inherits from another derived class, which in turn inherits from a base class.

class Grandparent:
    def grandparent_method(self):
        print("Grandparent's method")

class Parent(Grandparent):  # Parent inherits from Grandparent
    def parent_method(self):
        print("Parent's method")

class Child(Parent):  # Child inherits from Parent
    def child_method(self):
        print("Child's method")

child = Child()
child.grandparent_method() # Inherited from Grandparent
child.parent_method() # Inherited from Parent
child.child_method()

 4. Hierarchical Inheritance: Multiple derived classes inherit from the same base class.

class Shape:
    def area(self):
        print("Area of shape")

class Circle(Shape):  # Circle inherits from Shape
    pass

class Square(Shape):  # Square inherits from Shape
    pass

circle = Circle()
square = Square()
circle.area()
square.area()
