Q2. Difference between loc and iloc
loc[]: Label-based indexing.

iloc[]: Integer position-based indexing.

Example:

python
Copy
Edit
df.loc[1]     # Accesses row with index label 1
df.iloc[1]    # Accesses second row (0-based index)
Q3. Reindexing and Output of loc[2] and iloc[2]
python
Copy
Edit
reindex = [3, 0, 1, 2]
new_df = df.reindex(reindex)

print("new_df.loc[2]:\n", new_df.loc[2])
print("new_df.iloc[2]:\n", new_df.iloc[2])
new_df.loc[2]: Data from index label 2 (original 3rd row).

new_df.iloc[2]: Data at 3rd position of the reindexed frame (was originally at index 1).

Q8. Date difference calculator using pandas
python
Copy
Edit
import pandas as pd

date1 = pd.to_datetime(input("Enter first date (YYYY-MM-DD): "))
date2 = pd.to_datetime(input("Enter second date (YYYY-MM-DD): "))

delta = abs(date2 - date1)
print(f"Days: {delta.days}, Hours: {delta.total_seconds()//3600}, Minutes: {delta.total_seconds()//60}")
Observation:

delta.days gives whole days.

delta.total_seconds() gives exact difference including hours and minutes.

Edit

column_name = input("Enter column name to convert: ")

df[column_name] = pd.Categorical(df[column_name], categories=categories, ordered=True)

Q10. Visualize sales data using stacked bar chart
python
Copy
Edit
import pandas as pd
import matplotlib.pyplot as plt

file_path = input("Enter CSV file path: ")
df = pd.read_csv(file_path)

df_grouped = df.groupby(['Date', 'Category'])['Sales'].sum().unstack().fillna(0)
df_grouped.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title("Sales by Category Over Time")
plt.ylabel("Sales")
plt.xlabel("Date")
plt.legend(title="Category")
plt.show()
Q11. Student scores stats
python
Copy
Edit
import pandas as pd

file_path = input("Enter CSV file path: ")
df = pd.read_csv(file_path)

mean = df['Test Score'].mean()
median = df['Test Score'].median()
mode = df['Test Score'].mode().tolist()

print("+-----------+--------+")
print("| Statistic | Value  |")
print("+-----------+--------+")
print(f"| Mean      | {mean:.1f}   |")
print(f"| Median    | {median}    |")
print(f"| Mode      | {', '.join(map(str, mode))} |")
print("+-----------+--------+")
