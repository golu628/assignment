1. Five Common Seaborn Plots:

Scatter Plot: Visualizes the relationship between two numerical variables. Use it to explore potential correlations or patterns.
Line Plot: Shows trends of a numerical variable over time (or another categorical variable). Great for observing changes and forecasting.
Histogram: Depicts the distribution of a single numerical variable. Use it to understand data spread, skewness, and potential outliers.
Box Plot: Summarizes the distribution of a numerical variable with median, quartiles, and potential outliers. Useful for comparing distributions across categories.
# Violin Plot: Combines features of box plots and density plots, showing both the distribution shape and spread of a numerical variable across categories.

```python
#2.
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create the line plot with hue for different events
sns.lineplot(x="timepoint", y="signal", hue="event", data=fmri)
```

```python
#3.
import seaborn as sns

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Create two box plots side-by-side
sns.boxplot(
    x="pclass",
    y="age",
    showmeans=True,  # Show mean values as well
    data=titanic
)

sns.boxplot(
    x="pclass",
    y="fare",
    showmeans=True,  # Show mean values as well
    data=titanic
)
```

```python
#4.
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create a histogram with hue for different diamond cuts
sns.histplot(x="price", hue="cut", data=diamonds)
```

```python
#5.
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a pair plot with hue for different species
sns.pairplot(iris, hue="species")
```
