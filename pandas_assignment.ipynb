✅ Q2. Re-index DataFrame starting from 1 and increment by 2
python
Copy
Edit
def reindex_dataframe(df):
    df.index = range(1, 2*len(df)+1, 2)
    return df

# Example usage:
df = pd.DataFrame({'A': [5, 10], 'B': [15, 20], 'C': [25, 30]})
print(reindex_dataframe(df))

def add_word_count(df):
    df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split()))
    return df

# Example
df = pd.DataFrame({'Text': ['Hello world', 'Data Science Masters', 'Pandas is cool']})
print(add_word_count(df))

✅ Q7. Extract usernames from email addresses

def extract_usernames(df):
    df['Username'] = df['Email'].apply(lambda x: x.split('@')[0])
    return df

# Example
df = pd.DataFrame({'Email': ['john.doe@example.com', 'jane.smith@test.org']})
print(extract_usernames(df))

    mean = df['Values'].mean()
    median = df['Values'].median()
    std = df['Values'].std()
    return mean, median, std

def add_moving_average(df):
    df['MovingAverage'] = df['Sales'].rolling(window=7).mean()
    return df

# Example usage
df = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    'Sales': [100, 120, 130, 140, 150, 160, 170, 180, 190, 200]
})
print(add_moving_average(df))

def add_weekday(df):
    df['Date'] = pd.to_datetime(df['Date'])
    df['Weekday'] = df['Date'].dt.day_name()
    return df

# Example
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']})
print(add_weekday(df))
