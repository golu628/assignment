Q1: T-Test vs. Z-Test

T-Test: Used for comparing means of small samples (typically less than 30) or when the population standard deviation is unknown. It relies on the t-distribution.
Scenario: You want to compare the average weight gain of two different fertilizer treatments on a small group of plants (sample size < 30) and don't know the population standard deviation for weight gain.
Z-Test: Used for comparing means of large samples (typically greater than 30) where the population standard deviation is known. It relies on the standard normal distribution (z-distribution).
Scenario: You want to compare the average height of students in two different classes with a large sample size (n > 30) and have access to the standard deviation of heights in the entire student population.
Q2: One-Tailed vs. Two-Tailed Tests

One-Tailed Test: Used when you have a directional hypothesis about whether a mean or proportion will be greater than or less than a specific value.
Two-Tailed Test: Used when you have a non-directional hypothesis and want to see if the mean or proportion deviates from a specific value in either direction.
Q3: Type 1 and Type 2 Errors

Type 1 Error (False Positive): Rejecting a true null hypothesis (concluding a difference exists when there isn't).
Scenario: A drug is wrongly identified as effective when it has no real effect.
Type 2 Error (False Negative): Failing to reject a false null hypothesis (missing a real difference).
Scenario: A new teaching method shows no improvement over the existing one, but the test fails to detect a subtle positive effect.
Q4: Bayes' Theorem

Bayes' theorem helps calculate the probability of an event (B) occurring given the probability of another event (A) already happening and vice versa. It considers prior probability (initial belief about B) and likelihood (probability of observing A if B is true).

Example:

Probability of a disease (B) = 0.01 (prior probability)
Probability of a positive test (A) given the disease = 0.9 (likelihood)
Probability of a positive test (A) = 0.1 (including false positives)
What's the probability of having the disease given a positive test?

Solution:

Use the formula: P(B|A) = (P(A|B) * P(B)) / P(A)

Calculation:

CI = sample statistic ± (critical value * standard error)
